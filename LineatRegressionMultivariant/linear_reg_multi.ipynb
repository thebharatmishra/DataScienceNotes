# Linear Regression with Multiple Variable 

First we import all the necessary modules needed.

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

Now read the csv file using pandas and display it in tabular form

```python
df = pd.read_csv('homeprices.csv')
df
```

Now cleaning of data is necessary too so we to change the missing and NaN values to median of the column.

```python
import math

median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms =df.bedrooms.fillna(median_bedrooms)
df
```

Applying linear regression to the data we are able to determine the price of the the house. While taking in factor different variables such as area, bedrooms, age

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
reg.intercept_
reg.coef_

price_predict = reg.predict([[3000,3,40]])
print('Price(in $) is :',price_predict)
```
